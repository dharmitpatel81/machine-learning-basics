Gradient Descent

```python
""" 
- Gradient Descent is an algorithm that finds best fit line for given training dataset and it is trial and error based approach.

- In ML, We have to derive a relation from inputs and outputs.
    x = [1, 2, 3, 4, 5]
    y = [5, 7, 9, 11, 13]
    
    y = 2*x + 3 ==> (Prediction Function)

- MSE is also known as Cost Function.

- Slope of Straight line = Change in y / Change in x

- learning rate: In which speed we have to cut-down the step.

"""
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gradient_descent(x, y):
    m_curr = b_curr = 0
    iterations = 1000
    n = len(x)
    plt.scatter(x, y, color='red', marker='+', linewidths=5)
    learning_rate = 0.08

    for i in range(iterations):
        yp = m_curr * x + b_curr
        plt.plot(x, yp, color='green')

        # Mean Squared Error
        cost = (1 / n) * sum([val**2 for val in (y - yp)]) 

        # Partial Derivative of m and b
        md = -(2 / n) * sum(x * (y - yp))
        bd = -(2 / n) * sum(y - yp)

        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd
        # print("m {}, b {}, cost {} iterations {}".format(m_curr, b_curr, cost, i))
   
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 7, 9, 11, 13])
```

```python
gradient_descent(x, y)
```
